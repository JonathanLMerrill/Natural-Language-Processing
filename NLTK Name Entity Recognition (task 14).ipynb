```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('treebank')
nltk.download('words')
```

```python
sent = nltk.corpus.treebank.tagged_sents()[22]
print(sent)
print(nltk.ne_chunk(sent, binary=True))
```
